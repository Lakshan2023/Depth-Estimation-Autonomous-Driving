```python
import nbformat
import json

path = "Solution_2313081.ipynb"
nb = nbformat.read(path, as_version=4)

WIDGET_KEYS = {
    "application/vnd.jupyter.widget-view+json",
    "application/vnd.jupyter.widget-state+json"
}

removed = 0

def clean(obj):
    global removed

    if isinstance(obj, dict):
        # Remove widget MIME types
        for key in list(obj.keys()):
            if key in WIDGET_KEYS:
                obj.pop(key, None)
                removed += 1

        # Remove widget metadata blocks entirely
        if "widgets" in obj:
            obj.pop("widgets", None)
            removed += 1

        # Remove widget definitions missing 'state'
        if ("model_id" in obj) and ("state" not in obj):
            # erase entire invalid widget block
            obj.clear()
            removed += 1

        # Recurse into dict
        for key, value in list(obj.items()):
            clean(value)

    elif isinstance(obj, list):
        for item in obj:
            clean(item)

# Clean notebook root
clean(nb)

print(f"Total widget-related entries removed: {removed}")
nbformat.write(nb, path)
print("Notebook fully cleaned and saved.")
```
